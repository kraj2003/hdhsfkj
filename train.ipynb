```python
import pandas as pd

# -------------------------------
# 1. Load the IIS + Metrics Data
# -------------------------------
df_iis = pd.read_csv("./data/iis_logs_24.csv")       # your IIS log file
df_metrics = pd.read_csv("./data/iis_metrics_24.csv")  # your Metrics file
```

```python

# -------------------------------
# 2. Convert timestamp columns
# -------------------------------
df_iis["timestamp"] = pd.to_datetime(df_iis["timestamp"])
df_metrics["timestamp"] = pd.to_datetime(df_metrics["timestamp"])

# Sort both by timestamp (required for merge_asof)
df_iis = df_iis.sort_values("timestamp").reset_index(drop=True)
df_metrics = df_metrics.sort_values("timestamp").reset_index(drop=True)
```

```python
df_iis = df_iis[df_iis['timestamp'].dt.day != 24]
df_metrics = df_metrics[df_metrics['timestamp'].dt.day != 24]
```

```python
# Find exact match
# Target timestamp
target_ts = pd.to_datetime("2025-09-23 05:32:30")

idx = df_metrics.index[df_metrics['timestamp'] == target_ts]

if len(idx) > 0:
    # Found exact timestamp → get that + next 5 rows
    pos = idx[0]
    result = df_metrics.iloc[pos:pos+6]
else:
    # No exact match → find the next available timestamp
    next_idx = df_metrics[df_metrics['timestamp'] > target_ts].index.min()
    
    if pd.isna(next_idx):  # if no later data exists
        result = pd.DataFrame()
        print("No data found after the given timestamp.")
    else:
        pos = df_metrics.index.get_loc(next_idx)
        result = df_metrics.iloc[pos:pos+5]  # next 5 rows

result
```

```python

# -------------------------------
# 3. Merge IIS with nearest Metrics (CPU & RAM)
# -------------------------------
# df_final = pd.merge_asof(
#     df_iis,
#     df_metrics[["timestamp", "cpu_percent", "ram_percent"]],
#     on="timestamp",
#     direction="backward",    # use last known CPU/RAM values
#     tolerance=pd.Timedelta("30s")  # only merge if within 30 seconds
# )
df_final= pd.merge_asof(df_iis, df_metrics, on='timestamp', direction='nearest', tolerance=pd.Timedelta("5s"))
```

```python
df_final.head()
```

```python
df_final.shape
```

```python
# -------------------------------
# 4. Feature Engineering (to match your desired structure)
# -------------------------------

# If response_time > 2000 ms → delay_detected = 1 (delay happened)

# If response_time <= 2000 ms → delay_detected = 0 (no delay)

df_final['Delay_Detected'] = (df_final['response_time'] > 2000).astype(int)
# Break timestamp into date & time
df_final["date"] = df_final["timestamp"].dt.date
df_final["time"] = df_final["timestamp"].dt.time

# Extract hour & day of week
df_final["hour"] = df_final["timestamp"].dt.hour
df_final["day_of_week"] = df_final["timestamp"].dt.dayofweek

# Request path depth (count "/" in URL)
df_final["request_path_depth"] = df_final["url"].fillna("").str.count("/")

# Extract endpoint base (last part of URL path)
df_final["endpoint_base"] = df_final["url"].fillna("").str.split("/").str[-1]

# Service (just tagging IIS logs as 'api' here, adjust as needed)
df_final["service"] = "api"

# Flags for success/error
df_final["is_error"] = df_final["status_code"].apply(lambda x: 1 if x >= 400 else 0)
df_final["is_success"] = df_final["status_code"].apply(lambda x: 1 if 200 <= x < 400 else 0)

# Define peak hours (example: 9 AM - 6 PM)
df_final["is_peak_hour"] = df_final["hour"].apply(lambda h: 1 if 9 <= h < 18 else 0)

# Rename metrics columns to match final output format
df_final.rename(columns={"cpu_percent": "CPU", "ram_percent": "RAM"}, inplace=True)
```

```python
# -------------------------------
# 5. Reorder Columns (to match your expected output)
# -------------------------------

final_columns = [
    "date", "time", "server_ip", "method", "url", "site", "client_ip",
    "status_code", "response_time", "timestamp", "hour", "day_of_week",
    "request_path_depth", "endpoint_base", "service", "is_error",
    "is_success", "is_peak_hour", "CPU", "RAM"
]

df_final = df_final[[col for col in final_columns if col in df_final.columns]]

# -------------------------------
# 6. Save Final Dataset
# -------------------------------
# df_final.to_csv("./data/final_dataset.csv", index=False)

# print("✅ Final dataset created:", df_final.shape)
# print(df_final.head())
```

```python
df_5min['RAM'] = df_5min['RAM'].ffill().fillna(df_5min['RAM'].median())
# df_5min['CPU'] = df_5min['CPU'].ffill().fillna(df_5min['CPU'].median())

# # For response time, use forward fill then median
# df_5min['response_time'] = df_5min['response_time'].ffill().fillna(df_5min['response_time'].median())

df_5min["CPU"]=df_5min['CPU'].fillna(0)
df_5min["response_time"]=df_5min['response_time'].fillna(0)

# For categorical variables, use mode or 0
# df_5min['status_code'] = df_5min['status_code'].ffill().fillna(200)
df_5min['is_error'] = df_5min['is_error'].fillna(0)
df_5min['Delay_Detected'] = df_5min['Delay_Detected'].fillna(0)

# Create time-based features
df_5min['hour'] = df_5min.index.hour
df_5min['day_of_week'] = df_5min.index.dayofweek
df_5min['is_weekend'] = df_5min['day_of_week'].isin([5, 6]).astype(int)
# df_5min['is_peak_hour'] = df_5min['hour'].isin([9, 10, 11, 14, 15, 16]).astype(int)

# Define peak hours (example: 9 AM - 6 PM)
df_5min["is_peak_hour"] = df_5min["hour"].apply(lambda h: 1 if 9 <= h < 18 else 0)

# Create lag features for system health (NOT delay detection)
for lag in [1, 2, 3]:
    df_5min[f'cpu_lag_{lag}'] = df_5min['CPU'].shift(lag)
    df_5min[f'ram_lag_{lag}'] = df_5min['RAM'].shift(lag)
    df_5min[f'response_lag_{lag}'] = df_5min['response_time'].shift(lag)

# Rolling statistics for system health
df_5min['cpu_rolling_mean'] = df_5min['CPU'].rolling(window=6, min_periods=1).mean()
df_5min['cpu_rolling_std'] = df_5min['CPU'].rolling(window=6, min_periods=1).std()
df_5min['ram_rolling_mean'] = df_5min['RAM'].rolling(window=6, min_periods=1).mean()
df_5min['response_rolling_mean'] = df_5min['response_time'].rolling(window=6, min_periods=1).mean()

# System health indicators
df_5min['high_cpu'] = (df_5min['CPU'] > df_5min['CPU'].quantile(0.8)).astype(int)
df_5min['high_ram'] = (df_5min['RAM'] > df_5min['RAM'].quantile(0.8)).astype(int)
df_5min['high_error_period'] = (df_5min['is_error'] > 0).astype(int)

# Create proper target - predict delays 2 periods ahead WITHOUT using current delay
# This removes data leakage completely
df_5min['future_delay'] = df_5min['Delay_Detected'].shift(-2)
df_5min['target'] = df_5min['future_delay'].fillna(0).astype(int)

# Drop rows with NaN in lag features (first few rows)
# df_5min = df_5min.dropna()
```

```python
import pandas as pd

df1 = pd.read_csv("./data/final_dataset.csv")   # IIS / Apache style
df2 = pd.read_csv("./data/raw_data.csv")   # API log

# 2. Rename df1 columns to match df2 schema
rename_map_1 = {
    "cs_method": "method",
    "cs_uri_stem": "url",
    "s_ip": "site",
    "c_ip": "client_ip",
    "sc_status": "status_code",
    "time_taken": "response_time",
    "datetime": "timestamp"
}
df1 = df1.rename(columns=rename_map_1)

# 3. Define final schema
common_cols = [
    "date", "time", "method", "url", "site", "client_ip",
    "status_code", "response_time", "timestamp", "hour",
    "day_of_week", "request_path_depth", "endpoint_base",
    "service", "is_error", "is_success", "is_peak_hour",
    "CPU", "RAM"
]

# 4. Add missing columns (instead of KeyError)
for col in common_cols:
    if col not in df1.columns:
        df1[col] = pd.NA
    if col not in df2.columns:
        df2[col] = pd.NA

# 5. Reorder both
df1 = df1[common_cols]
df2 = df2[common_cols]

# 6. Concatenate vertically
df_merged = pd.concat([df1, df2], ignore_index=True)

# 7. Convert timestamp safely
df_merged["timestamp"] = pd.to_datetime(df_merged["timestamp"], errors="coerce")

print("Final merged DataFrame shape:", df_merged.shape)
print(df_merged.head())
```

```python
import pandas as pd

# ---------------------------
# Load your two datasets
# ---------------------------
df1 = pd.read_csv("./data/final_dataset.csv")   # IIS / Apache style
df2 = pd.read_csv("./data/raw_data.csv")


# ---------------------------
# Define a common schema mapping
# ---------------------------
column_mapping = {
    # df1 → df2 names
    'cs_method': 'method',
    'cs_uri_stem': 'url',
    's_port': 'port',
    'c_ip': 'client_ip',
    'sc_status': 'status_code',
    'sc_substatus': 'sub_status',
    'sc_win32_status': 'win32_status',
    'time_taken': 'response_time',
    'datetime': 'timestamp',
    's_ip': 'server_ip',
    'site': 'site',   # already same
    'service': 'service'
}

# Apply mapping to df1
df1 = df1.rename(columns=column_mapping)

# ---------------------------
# Make sure both DataFrames have the same column order
# (take union of all columns so nothing is lost)
# ---------------------------
all_columns = list(set(df1.columns).union(set(df2.columns)))
df1 = df1.reindex(columns=all_columns)
df2 = df2.reindex(columns=all_columns)

# ---------------------------
# Merge (stack vertically, no loss)
# ---------------------------
df_merged = pd.concat([df1, df2], ignore_index=True)

# ---------------------------
# Ensure timestamp is datetime
# ---------------------------
df_merged['timestamp'] = pd.to_datetime(df_merged['timestamp'], errors='coerce')

# ---------------------------
# Create Delay_Detected
# ---------------------------
df_merged['Delay_Detected'] = (df_merged['response_time'] > 2000).astype(int)

# ---------------------------
# Print summary
# ---------------------------
print(f"Original data shape: {df_merged.shape}")
print(f"Delay instances: {df_merged['Delay_Detected'].sum()}/{len(df_merged)} "
      f"({df_merged['Delay_Detected'].mean()*100:.1f}%)")
```
